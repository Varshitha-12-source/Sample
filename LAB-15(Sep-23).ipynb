```python
import numpy as np
import matplotlib.pyplot as plt

# Create a new dataset with different characteristics
data = {
    'a': np.linspace(0, 10, 50),                      # x-values from 0 to 10 (evenly spaced)
    'c': np.random.randint(10, 100, 50),              # random color values from 10 to 99
    'd': np.random.uniform(0.5, 2.0, 50)              # random sizes between 0.5 and 2.0
}

# y-values: quadratic trend with some noise
data['b'] = data['a']**2 + np.random.normal(0, 5, 50)

# Scale 'd' for use as marker sizes in scatter plot
data['d'] = data['d'] * 100                          # scale up sizes for visibility

# Create scatter plot using dictionary keys as column names
plt.scatter('a', 'b', c='c', s='d', data=data, cmap='viridis', alpha=0.7, edgecolors='w')

# Label axes
plt.xlabel('entry a (X values)')
plt.ylabel('entry b (Y values)')

# Add a colorbar for reference to 'c' values
plt.colorbar(label='Color scale (c)')

# Show the plot
plt.title('Scatter Plot with Varying Size and Color')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the damped cosine function
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

# Create two time arrays with different resolutions
t1 = np.arange(0.0, 5.0, 0.1)   # Coarse time intervals
t2 = np.arange(0.0, 5.0, 0.02)  # Finer time intervals

# Create a new figure
plt.figure(figsize=(8, 6))  # Optional: set figure size

# --- First Subplot: Damped Cosine ---
plt.subplot(2, 1, 1)  # (nrows=2, ncols=1, index=1)
plt.plot(t1, f(t1), 'bo', label='Sampled')    # Blue circles
plt.plot(t2, f(t2), 'k', label='Smooth')      # Black line
plt.title('Damped Cosine Wave')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)

# --- Second Subplot: Pure Cosine ---
plt.subplot(2, 1, 2)  # (nrows=2, ncols=1, index=2)
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')   # Red dashed line
plt.title('Pure Cosine Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# Automatically adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

```python
def rgb_to_grayscale(img):
    if len(img.shape) == 3:
        return np.dot(img[...,:3], [0.299, 0.587, 0.114]).astype(np.uint8)
    return img

gray_bug = rgb_to_grayscale(bug_img)

plt.imshow(gray_bug, cmap="gray")
plt.title(" Grayscale Bug Image")
plt.axis("off")
plt.show()
```

```python
from PIL import ImageFilter

# Convert bug_img to PIL Image, apply Gaussian blur, convert back to numpy array
blurred_bug = np.asarray(Image.fromarray(bug_img).filter(ImageFilter.GaussianBlur(radius=5)))

# Display original image
plt.subplot(1, 2, 1)
plt.imshow(bug_img)
plt.title("Original Bug")
plt.axis("off")

# Display blurred image
plt.subplot(1, 2, 2)
plt.imshow(blurred_bug)
plt.title("Blurred Bug")
plt.axis("off")

plt.show()  # Show both images side by side
```

```python
def apply_color_filter(img, color_filter):
    filtered = img.copy().astype(float)  # Copy image and convert to float for scaling
    if len(img.shape) == 3 and img.shape[2] >= 3:  # Check for RGB image
        filtered[..., 0] *= color_filter[0]  # Scale Red channel
        filtered[..., 1] *= color_filter[1]  # Scale Green channel
        filtered[..., 2] *= color_filter[2]  # Scale Blue channel
    # Clip values to [0,255] and convert back to uint8
    return np.clip(filtered, 0, 255).astype(np.uint8)

# Apply different color filters to emphasize red, green, or blue
red_bug   = apply_color_filter(bug_img, [1.5, 0.5, 0.5])   # Increase red
green_bug = apply_color_filter(bug_img, [0.5, 1.5, 0.5])   # Increase green
blue_bug  = apply_color_filter(bug_img, [0.5, 0.5, 1.5])   # Increase blue

# Plot filtered images side-by-side
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(red_bug)
plt.title("Red Filter")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(green_bug)
plt.title("Green Filter")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(blue_bug)
plt.title("Blue Filter")
plt.axis("off")

plt.show()
```

```python
from sklearn.cluster import KMeans

# Reshape image to (num_pixels, 3)
pixels = bug_img.reshape(-1, 3)

# Apply KMeans to reduce to 4 colors
kmeans = KMeans(n_clusters=4, random_state=0).fit(pixels)
quantized = kmeans.cluster_centers_[kmeans.labels_]
quantized_img = quantized.reshape(bug_img.shape).astype(np.uint8)

# Show quantized image
plt.figure()
plt.imshow(quantized_img)
plt.title("Color Quantization (4 Colors)")
plt.axis("off")
plt.show()
```
