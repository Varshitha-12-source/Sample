# HANDLING JSON FILE IN PYTHON 

# Write Python Data to JSON File 

```python
import json

def write_to_json(data, filename="output.json"):
    """
    Write Python data to a JSON file with proper formatting
    
    Parameters:
    data (dict/list): Python dictionary or list to be converted to JSON
    filename (str): Name of the output JSON file (default: "output.json")
    
    Returns:
    None: Writes data to file and prints status message
    """
    try:
        # Open the file in write mode
        with open(filename, 'w') as json_file:
            # Convert Python object to JSON and write to file
            # indent=4 adds pretty formatting with 4 spaces
            json.dump(data, json_file, indent=4)
        print(f" Data successfully written to {filename}")
    
    except PermissionError:
        print(f" Permission denied: Cannot write to {filename}")
    except Exception as e:
        print(f" Error writing to JSON file: {e}")

# Example usage with sample data
sample_data = {
    "name": "John Doe",
    "age": 30,
    "city": "New York",
    "codes": ["PY001", "JS002", "HTML003"],
    "is_active": True,
    "contact": {
        "email": "john.doe@example.com",
        "phone": "+1234567890"
    }
}

# Write the sample data to a JSON file
write_to_json(sample_data, "output.json")

# Also convert to JSON string format for display/API usage
json_string = json.dumps(sample_data, indent=2)
print("\n JSON String Format:")
print(json_string)

# Additional example: Writing list data
list_data = [
    {"id": 1, "name": "Item 1", "code": "ITM001"},
    {"id": 2, "name": "Item 2", "code": "ITM002"},
    {"id": 3, "name": "Item 3", "code": "ITM003"}
]

write_to_json(list_data, "list_output.json")
print("\n List data written to list_output.json")
```

```python
import json

def parse_json_search_clear(filename, search_code=None, clear_codes=False):
    try:
        # Read and parse the JSON file
        with open(filename, 'r') as json_file:
            data = json.load(json_file)
        
        print(f" Successfully parsed {filename}")
        print(" Original Data Structure:")
        print(json.dumps(data, indent=2))
        
        # Search for specific code if provided
        if search_code:
            found = search_for_code(data, search_code)
            if found:
                print(f"\nCode '{search_code}' found in the data!")
            else:
                print(f"\n Code '{search_code}' not found in the data!")
        
        # Clear all code fields if requested
        if clear_codes:
            cleaned_data = clear_code_fields(data)
            
            # Write the cleaned data back to file
            with open(filename, 'w') as json_file:
                json.dump(cleaned_data, json_file, indent=4)
            
            print(f"\n All code fields cleared from {filename}")
            print(" Updated Data Structure:")
            print(json.dumps(cleaned_data, indent=2))
            
            return cleaned_data
        
        return data
        
    except FileNotFoundError:
        print(f" Error: File '{filename}' not found!")
    except json.JSONDecodeError:
        print(f" Error: Invalid JSON format in '{filename}'!")
    except Exception as e:
        print(f" Unexpected error: {e}")

def search_for_code(data, target_code):
    """
    Recursively search for a code value in JSON data
    
    Parameters:
    data: JSON data (dict, list, or primitive)
    target_code (str): Code value to search for
    
    Returns:
    bool: True if code found, False otherwise
    """
    # If data is a dictionary, check each key-value pair
    if isinstance(data, dict):
        for key, value in data.items():
            # Check if current key contains 'code' (case-insensitive)
            if 'code' in key.lower():
                if isinstance(value, list) and target_code in value:
                    return True
                elif value == target_code:
                    return True
            # Recursively search nested structures
            if isinstance(value, (dict, list)):
                if search_for_code(value, target_code):
                    return True
    
    # If data is a list, search each item
    elif isinstance(data, list):
        for item in data:
            if search_for_code(item, target_code):
                return True
    
    return False

def clear_code_fields(data):
    """
    Recursively remove all code-related fields from JSON data
    
    Parameters:
    data: JSON data to clean
    
    Returns:
    dict/list: Data with code fields removed
    """
    # Handle dictionary data
    if isinstance(data, dict):
        cleaned_dict = {}
        for key, value in data.items():
            # Skip fields that contain 'code' in their name (case-insensitive)
            if 'code' not in key.lower():
                # Recursively clean nested structures
                if isinstance(value, (dict, list)):
                    cleaned_dict[key] = clear_code_fields(value)
                else:
                    cleaned_dict[key] = value
        return cleaned_dict
    
    # Handle list data
    elif isinstance(data, list):
        return [clear_code_fields(item) for item in data]
    
    # Return primitive values unchanged
    else:
        return data

# Example usage of the parsing and clearing function
if __name__ == "__main__":
    # First, let's create a sample JSON file to work with
    sample_data = {
        "user": {
            "name": "Alice Johnson",
            "user_code": "USR001",
            "access_codes": ["READ", "WRITE", "EXECUTE"],
            "profile": {
                "level_code": "PRO",
                "settings": {"theme": "dark", "language": "en"}
            }
        },
        "projects": [
            {
                "name": "Website Redesign",
                "project_code": "PRJ001",
                "status": "active",
                "tags": ["web", "react", "python"]
            },
            {
                "name": "Mobile App",
                "project_code": "PRJ002", 
                "status": "completed",
                "tags": ["mobile", "flutter"]
            }
        ],
        "system_codes": ["SYS_A", "SYS_B", "SYS_C"]
    }
    
    # Create the sample file
    with open("sample_data.json", 'w') as f:
        json.dump(sample_data, f, indent=2)
    
    print("===  Searching for Code 'PRJ001' ===")
    parse_json_search_clear("sample_data.json", search_code="PRJ001")
    
    print("\n===  Searching for Code 'NONEXISTENT' ===")
    parse_json_search_clear("sample_data.json", search_code="NONEXISTENT")
    
    print("\n===  Clearing All Codes ===")
    cleaned_data = parse_json_search_clear("sample_data.json", clear_codes=True)
    
    print("\n===  Verifying Codes are Removed ===")
    parse_json_search_clear("sample_data.json", search_code="PRJ001")
```
